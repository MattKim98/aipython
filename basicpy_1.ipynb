```python
#print(a+b) #형식 2+1=3
print(f'{a}+{b}={a+b}') #형식 2+1=3
#파이썬에서 문자열 앞의 f는 **f-string(Formatted String Literals)**을 의미합니다.

#f-string의 주요 용도는 다음과 같습니다:

#변수 삽입: 중괄호 { } 안에 변수 이름을 넣으면 그 값이 문자열에 직접 포함됩니다.
#직관적인 가독성: format() 함수나 % 연산자를 사용하는 것보다 코드가 훨씬 간결하고 읽기 쉽습니다.
#식 계산: {a+b}처럼 중괄호 안에서 직접 연산을 수행할 수도 있습니다.
#예를 들어, print(f'{a}+{b}={a+b}')는 변수 a, b의 값과 그 합을 계산하여 하나의 문자열로 만들어 출력해 줍니다.
```

```python
# 1. 변수 선언 (숫자 저장)
num1 = 10
num2 = 3

# 2. 사칙연산 연산자를 사용한 계산
plus = num1 + num2
minus = num1 - num2
multiply = num1 * num2
divide = num1 / num2

# 3. f-string 없이 결과 출력 (쉼표 사용)
print(num1, '+', num2, '=', plus)
print(num1, '-', num2, '=', minus)
print(num1, '*', num2, '=', multiply)
print(num1, '/', num2, '=', round(divide, 2)) # 소수점 둘째자리까지 반올림

# 4. 결과를 텍스트 파일로 저장
file_name = "output.txt"
with open(file_name, 'w', encoding='utf-8') as f:
    f.write(f"{num1} + {num2} = {plus}\n")
    f.write(f"{num1} - {num2} = {minus}\n")
    f.write(f"{num1} * {num2} = {multiply}\n")
    f.write(f"{num1} / {num2} = {round(divide, 2)}\n")

print(f"\n결과가 '{file_name}' 파일에 저장되었습니다.")
```
