```python
import random
answer = random.randint(1, 100)
user = 0
count = 1
user = int(input('숫자 입력'))
while user != answer:
    if user < answer:
            print("더 큰 수를 입력하시오")
            user = int(input('숫자 입력'))
            count +=1
    elif user > answer:
            print("더 작은 수를 입력하시오")
            user = int(input('숫자 입력'))
            count +=1
print("정답")
print(f"총 시도횟수는 {count}")
```

```python
num = int(input("약수를 구할 숫자를 입력하세요: "))

print(f"{num}의 약수:")
for i in range(1, num + 1):
    if num % i == 0:
        print(i)
```
