```python
height1 = float(input("학생 1의 키(cm)를 입력하시오"))
height2 = float(input("학생 2의 키(cm)를 입력하시오"))
if height1 > height2:
    print("첫번째 학생이 더 크다")
else:
    print("두번째 학생이 더 크다")
```

```python
height1 = float(input("학생 1의 키(cm)를 입력하시오"))
height2 = float(input("학생 2의 키(cm)를 입력하시오"))
if height1 == height2:
    print("두 학생의 키는 같다")
elif height1 . height2:
    print("첫번째 학생이 더 크다")
else:
    print("두번째 학생이 더 크다")
```

```python
#하나의 점수를 입력받음, 90점이상 A, 80점이상 B, 70점이상 C, 나머지 D를 출력
score = int(input("점수를 입력하시오"))
if score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
else:
    print("D")
```

```python
import random
a = random.randint(1,3)
b = random.randint(1,3)
if a == b:
  print(a,b)
  print('무승부')
elif (a==1 and b==3) or (a==2 and b==1) or (a==3 and b==2): # A 승리 조건 (가위>보, 바위>가위, 보>바위)
  print ('A 승리')
else
  print('B 승리')
```
