```python
fruits = ["apple", "banana", "cherry"]
```

```python
fruits.extend([
    "orange", "grape", "mango", "pineapple", "strawberry",
    "watermelon", "peach", "kiwi", "blueberry", "lemon"
])
```

```python
len(fruits)
```

```python
for fruit in fruits:
    print(fruit)
```

```python
result = []

for fruit in fruits:
    result.append(fruit.upper())

print(result)
```

```python
new_list = []

for fruit in fruits:
    if len(fruit) >= 6:
        new_list.append(fruit)

print(new_list)
```

```python
for i, fruit in enumerate(fruits):
    print(i+1, fruit)
```

```python
joseon = [
    "태조", "정종", "태종", "세종", "문종", "단종",
    "세조", "예종", "성종", "연산군", "중종", "명종",
    "선조", "광해군", "인조", "효종", "현종", "숙종",
    "경종", "영조", "정조", "순조", "헌종", "철종",
    "고종", "순종"
]
```

```python
for i, king in enumerate(joseon, start=1):
    print(f"{i}대 왕은 {king}입니다")
```

```python
n = int(input("몇 번째 왕? "))

if 1 <= n <= len(joseon):
    print(f"{n}대 왕은 {joseon[n-1]}입니다")
else:
    print("없음")
```

```python
name = input("왕 이름을 입력하세요: ")

if name in joseon:
    index = joseon.index(name)
    print(f"{name}은 {index + 1}번째 왕입니다")
else:
    print("해당 왕이 없습니다")
```

```python
rev = reversed(joseon)
for rev in rev:
    print(rev, end=" ")
```
