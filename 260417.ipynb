```python
for key, value in cafe_menu.items():
  print(f"{key}의 가격은 {value}원입니다.")
```

```python
total = 0
for val in cafe_menu.values():
  total += val
print(f"총 가격은 {total}원 입니다")
```

```python
total_order_price = 0
print("어서오세요! 메뉴를 선택하거나 '종료'를 입력하여 주문을 완료하세요.")

print("\n--- 메뉴 ---")
for item, price in cafe_menu.items():
    print(f"{item}: {price}원")
print("-----------------")

while True:
    order = input("주문할 메뉴를 입력하세요 (종료하려면 '종료' 입력): ")

    if order == '종료':
        print(f"\n주문이 완료되었습니다. 총 가격은 {total_order_price}원 입니다.")
        break
    elif order in cafe_menu:
        price = cafe_menu[order]
        total_order_price += price
        print(f"{order} {price}원이 추가되었습니다. 현재 총 가격: {total_order_price}원")
    else:
        print("잘못된 메뉴입니다. 다시 선택해주세요.")
```
