```python
# 분 입력받기
a = int(input("분을 입력하세요: "))

# 시간과 분 계산
b = a // 60
c = a % 60

# 결과 출력
print(f"{a}분은 {b}시간 {c}분입니다")
```

```python
cafe_price = 2000
latte_price = 3000
greentea_price = 2500
juice_price = 2800
print('커피 2000원\n라떼 3000원\n녹차 2500원\n주스 2800원')
cafe_count = int(input('커피 주문수'))
latte_count = int(input('라떼 주문수'))
greentea_count = int(input('녹차 주문수'))
juice_count = int(input('주스 주문수'))
total_price = cafe_price * cafe_count + latte_price * latte_count + greentea_price * greentea_count + juice_price * juice_count
print(f'총 금액은 {total_price}원 입니다')
```
